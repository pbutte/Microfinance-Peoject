```python
#setting features for the household which are connected to grid and not connected to grid
df.loc[df["Connected to grid"] == "Yes", ["Why not connected to grid?", "Electric meter"]] = ["Already Connected", "Yes"]
df.loc[df["Connected to grid"] == "No", ["Electric meter"]] = ["No"]

#households which cannot to grid can connect to the grid by illegal way
df.loc[df["Why not connected to grid?"] == "Connection cost is very high", ["Who do you pay", "Electric meter"]]= ["No one (for free)", "No"]
df.loc[df["Why not connected to grid?"] == "Can't afford electricity bill", ["Who do you pay", "Electric meter"]]= ["No one (for free)", "No"]


#legality check
df.loc[df["Who do you pay"] == "No one (for free)", ["Electric meter"]] = ["No"]

#each village would be having its own electricity supply
df.loc[df["Village"]== "Village 1", ["Electricity available per day(in hrs)"]]= [random.randint(2,24)]
df.loc[df["Village"]== "Village 2", ["Electricity available per day(in hrs)"]]= [random.randint(2,24)]
df.loc[df["Village"]== "Village 3", ["Electricity available per day(in hrs)"]]= [random.randint(2,24)]
df.loc[df["Village"]== "Village 4", ["Electricity available per day(in hrs)"]]= [random.randint(2,24)]

#each village would be having its own power cuts
df.loc[df["Village"]== "Village 1", ["Power cuts per week"]]= [random.randint(2,20)]
df.loc[df["Village"]== "Village 2", ["Power cuts per week"]]= [random.randint(2,20)]
df.loc[df["Village"]== "Village 3", ["Power cuts per week"]]= [random.randint(2,20)]
df.loc[df["Village"]== "Village 4", ["Power cuts per week"]]= [random.randint(2,20)]

#minimum temperature in winter
df.loc[df["Village"]== "Village 1", ["Minimum temperature in Winter"]]= 10
df.loc[df["Village"]== "Village 2", ["Minimum temperature in Winter"]]= 10
df.loc[df["Village"]== "Village 3", ["Minimum temperature in Winter"]]= 10
df.loc[df["Village"]== "Village 4", ["Minimum temperature in Winter"]]= 10

#maximum temperature in summer
df.loc[df["Village"]== "Village 1", ["Maximum temperature in Summer"]]= 42
df.loc[df["Village"]== "Village 2", ["Maximum temperature in Summer"]]= 42
df.loc[df["Village"]== "Village 3", ["Maximum temperature in Summer"]]= 42
df.loc[df["Village"]== "Village 4", ["Maximum temperature in Summer"]]= 42

#rainy season
df.loc[df["Village"]== "Village 1", ["How many months does it rain"]]= 4
df.loc[df["Village"]== "Village 2", ["How many months does it rain"]]= 4
df.loc[df["Village"]== "Village 3", ["How many months does it rain"]]= 4
df.loc[df["Village"]== "Village 4", ["How many months does it rain"]]= 4

#household having battries where electrcity is available for less than 8 hours
def battery(b):
    if 0<=b<=8: batteries= "Yes"
    else: batteries= "No"
    return(batteries)
df["Rechargeable batteries"]= df["Electricity available per day(in hrs)"].apply(battery)
df.loc[df["Rechargeable batteries"] != "Yes", ["Use of batteries", "Cost of rechargeable battery", "Electricity source to recharge the battery"]] = ["NA", "NA", "NA"]

#monthly income is defined on number of family members in a family
def income(i):
    if 2<=i<4: money= random.randint(3000, 5000)
    elif 4<=i<=6: money= random.randint(5001, 10000)
    else: money= random.randint(10001, 15000)
    return(money)
df["Monthly Income"]= df["Family Members"].apply(income)

#solar devices can be purchased by the households having monthly income between Rs. 12000 and Rs. 15000
def solar(s):
    if 12000 <= s <= 15000: solardevice= "Yes"
    else: solardevice= "No"
    return(solardevice)
df["Solar devices"]= df["Monthly Income"].apply(solar)

#reasons for not having solar devices
reasons= ["Initial investment is high",  "No subsidies are provided"]
def nosolar(s1):
    if s1 <= 5000: reason= "Low income"
    else: reason= random.choice(reasons)
    return(reason)
df["Why no solar?"]= df["Monthly Income"].apply(nosolar)

#other featues dependant on solar devices
df.loc[df["Solar devices"] == "Yes" , ["Why no Solar?", "Willing to shift to solar applicanes"]]= ["Already using", "Already using"]
df.loc[df["Solar devices"] == "No", ["Solar Usage (Months)", "Cost of Solar", "Savings after installing a solar light"]]= ["Never Used", "Never Purchased", "Never installed"]

#batttries are charged through solar when a household has already install solar 
df.loc[(df["Solar devices"] == "Yes") & (df["Rechargeable batteries"] == "Yes"), "Electricity source to recharge the battery"] = "Solar"

#kerosene and other lightning fuels are used by the households which are earning less then 6000 and having electricity supply for less than 8 hours
df["Did you use KEROSENE/FUEL-BASED /CANDLE LIGHTING"]= np.where((df["Monthly Income"] <= 6000) & (df["Electricity available per day(in hrs)"] <= 8), "Yes", "No")
df.loc[df["Did you use KEROSENE/FUEL-BASED /CANDLE LIGHTING"] == "Yes", ["Fuel source for LAMP", "How often the [FUEL TYPE] available"]]= ["NA", "NA"]

#electrical appliances are based on number of members in particular family
def electricalappliances(e):
    if 1<=e<=2: appliances= 3
    elif e == 3: appliances= 4
    elif 4<=e<=5: appliances= 5
    elif 6<=e<=7: appliances= 7
    else: appliances= 9
    return(appliances)
df["Electrical appliances"]= df["Family Members"].apply(electricalappliances)

#other featues dependent on no grid connection
df.loc[df["Why not connected to grid?"] == "No grid near by", ["Electrical appliances", "Electricity available per day(in hrs)", "Electric meter", "Electric bill of last month", "Power cuts per week", "Voltage problems if any", "Worst months for service from the grid", "Household members affected", "Serious problems you experience with grid electricity", "How often you pay for electricity bill", "Who do you pay", "How do you make your payment", "Household members die because of the grid electricity", "Rechargeable batteries", "Electricity source to recharge the battery"]]= [0, 0, "No", 0, 0, "Not connected to grid", "Not connected to grid", "Not connected to grid", "Not connected to grid", "Not connected to grid", "Not connected to grid", "Not connected to grid", "Not connected to grid", "Yes", "Shop"]

#usage of mobile charger according to electricity available ans its consumption for particular time
def mobile(m):
    if 1<m<=3: usage= 10
    elif 4<=m<=7: usage= 15
    elif 8<=m<=11: usage= 20
    elif 12<=m<=15: usage= 25
    elif 16<=m<=22: usage= 30
    else: usage= 30
    return(usage)
df["Mobile Charger(Wh)"]= df["Electricity available per day(in hrs)"].apply(mobile)

#usage of 5W LED Torch according to electricity available and its consumption for particular time
def torch5W(t):
    if 1<t<=3: usage= 5
    else: usage= 10
    return(usage)
df["5W LED Torch"]= df["Electricity available per day(in hrs)"].apply(torch5W)
        
#usage of 10W LED Torch according to electricity available and its consumption for particular time
def torch10W(t):
    if 1<t<=3: usage= 10
    else: usage= 20
    return(usage)
df["10W LED Torch"]= df["Electricity available per day(in hrs)"].apply(torch10W)

#usage of 5W LED bulb according to electricity available and its consumption for particular time
def led5W(l):
    if 1<l<=3: usage= 10
    elif 4<=l<=7: usage= 20
    elif 8<=l<=11: usage= 30
    elif 12<=l<=15: usage= 40
    elif 16<=l<=22: usage= 50
    else: usage= 55
    return(usage)
df["5W LED Bulb(Wh)"]= df["Electricity available per day(in hrs)"].apply(led5W)

#usage of 10W LED bulb according to electricity available and its consumption for particular time
def led10W(lo):
    if 1<lo<=3: usage= 20
    elif 4<=lo<=7: usage= 40
    elif 8<=lo<=11: usage= 60
    elif 12<=lo<=15: usage= 60
    elif 16<=lo<=22: usage= 80
    else: usage= 90
    return(usage)
df["10W LED Bulb(Wh)"]= df["Electricity available per day(in hrs)"].apply(led10W)

#usage of 45W Fan according to electricity available and its consumption for particular time
def fan(f):
    if 1<f<=3: usage= 90
    elif 4<=f<=7: usage= 180
    elif 8<=f<=11: usage= 360
    elif 12<=f<=15: usage= 450
    elif 16<=f<=22: usage= 540
    else: usage= 540
    return(usage)
df["45W Fan(Wh)"]= df["Electricity available per day(in hrs)"].apply(fan)

#usage of 55W TV according to electricity available and its consumption for particular time
def TV(t):
    if 1<t<=3: usage= 110
    elif 4<=t<=7: usage= 220
    elif 8<=t<=11: usage= 330
    elif 12<=t<=15: usage= 440
    elif 16<=t<=22: usage= 440
    else: usage= 440
    return(usage)
df["55W TV(Wh)"]= df["Electricity available per day(in hrs)"].apply(TV)

#usage of 130W Fridge according to electricity available
df["130W Fridge(Wh)"]= df["Electricity available per day(in hrs)"]*130

#selection of electrical applicanes 
df.loc[df["Electrical appliances"] == 0, ["Mobile Charger(Wh)", "5W LED Torch", "10W LED Torch", "5W LED Bulb(Wh)", "10W LED Bulb(Wh)", "45W Fan(Wh)", "55W TV(Wh)", "130W Fridge(Wh)"]]= [0, 0, 0, 0, 0, 0, 0, 0]
df.loc[df["Electrical appliances"] == 3, ["10W LED Torch", "10W LED Bulb(Wh)", "45W Fan(Wh)", "55W TV(Wh)", "130W Fridge(Wh)"]] = [0, 0, 0, 0, 0]
df.loc[df["Electrical appliances"] == 4, ["10W LED Torch", "45W Fan(Wh)", "55W TV(Wh)", "130W Fridge(Wh)"]] = [0, 0, 0, 0]
df.loc[df["Electrical appliances"] == 5, ["5W LED Torch", "55W TV(Wh)", "130W Fridge(Wh)"]] = [0, 0, 0]
df.loc[df["Electrical appliances"] == 7, ["5W LED Torch", "130W Fridge(Wh)"]] = [0, 0]
df.loc[df["Electrical appliances"] == 9, ["5W LED Torch"]] = [0]

df["5W LED Bulb(Wh)"]= np.where(df["Electrical appliances"] == 7, df["5W LED Bulb(Wh)"]*2, df["5W LED Bulb(Wh)"])
df["5W LED Bulb(Wh)"]= np.where(df["Electrical appliances"] == 9, df["5W LED Bulb(Wh)"]*2, df["5W LED Bulb(Wh)"])
df["10W LED Torch"]= np.where(df["Electrical appliances"] == 9, df["10W LED Torch"]*2, df["10W LED Torch"])

#total capacity of a household(Wh)
df["Capacity(Wh)"]= df["Mobile Charger(Wh)"]+df["5W LED Torch"]+df["10W LED Torch"]+df["5W LED Bulb(Wh)"] + df["10W LED Bulb(Wh)"] + df["45W Fan(Wh)"] + df["55W TV(Wh)"]+ df["130W Fridge(Wh)"]
#df["Capacity(Wh) Monthly"]= df["Capacity(Wh)"]*30

df["Units Monthly"]= df["Capacity(Wh)"]*30/1000
    

#monthly electricity bill(100- fixed charge, 3.46/unit(1-100 units) and 7.43/unit(100 above units), 1.45 per unit- carrying charges, 16% electricty charges(fixed charge+ units charges+carrying charges))
def bill(b):
    if b == 0: pay= 0
    elif 1<=b<=100: pay=100+b*3.46+b*1.45+(100+b*3.46+b*1.45)*0.16
    else: pay= 90+b*7.43+b*1.45+(100+b*3.48+b*1.45)*0.16
    return(pay)
df["Electric bill of last month"]= df["Units Monthly"].apply(bill)

#capacity tier ranking based on capacity of particular household
def captier(n):
    if 0<=n<=11: tier= 0
    elif 12<=n<=199: tier= 1
    elif 200<=n<=999: tier= 2
    elif 1000<=n<=3399: tier= 3
    elif 3400<=n<= 8199: tier= 4
    else: tier= 5
    return(tier)
df["Capacity Tier"]= df["Capacity(Wh)"].apply(captier)

#availabilty tier ranking based on electricity avaliable for particular household
def availabilitytier(a):
    if 0<=a<=3: tier= 0
    elif 4<=a<=7: tier= 2
    elif 8<=a<=15: tier= 3
    elif 16<=a<=22: tier= 4
    else: tier= 5
    return(tier)
df["Availabilty Tier"]= df["Electricity available per day(in hrs)"].apply(availabilitytier)

#reliability tier ranking based on number of power cuts a particular household experience
def reliabilitytier(r):
    if 0<=r<=3: tier= 5
    elif 4<=r<=13: tier= 4
    else: tier= 3
    return(tier)
df["Reliability Tier"]= df["Power cuts per week"].apply(reliabilitytier)

#quality tier ranking based on voltages problems due to which the electrical appliances got affected
df["Quality Tier"]= df["Voltage problems if any"].apply(lambda x: 5 if x == "No Problem" else 3)

#affordability tier ranking based on annual monthly income and total annual expenses of energy (365kWh)
def affordabilitytier(af):
    if af*0.05*12 > 3766.8: tier= 5
    else: tier= 2
    return(tier)
df["Affordability Tier"]= df["Monthly Income"].apply(affordabilitytier)

#legality tier ranking based on payment mode of a particular household
df["Legality Tier"]= df["Who do you pay"].apply(lambda x: 3 if x == "No one (for free)" else 5)

#health & safety  tier ranking based on the casuality experienced by the household
def health(h):
    if h== "Yes": tier= 3
    elif h== "No": tier= 5
    else: tier= 3
    return(tier)
df["Health & Safety Tier"]= df["Household members die because of the grid electricity"].apply(health)

#tier ranking of individual household- selection of lowest tier based on the 7 tier rankings for particular attribute
df["Tier Ranking"]= df.loc[:, ["Capacity Tier", "Availabilty Tier", "Reliability Tier", "Quality Tier", "Affordability Tier", "Legality Tier", "Health & Safety Tier"]].min(axis= 1)
df
```
