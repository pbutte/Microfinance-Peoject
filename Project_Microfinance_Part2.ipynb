```python
from sklearn import preprocessing
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
```

```python
scaler = MinMaxScaler()

scaler.fit(df[["Availabilty Tier"]])
df["Availabilty Tier"] = scaler.transform(df[["Availabilty Tier"]])

scaler.fit(df[["Electricity available per day(in hrs)"]])
df["Electricity available per day(in hrs)"] = scaler.transform(df[["Electricity available per day(in hrs)"]])
```

```python
#Using the elbow method to find the optimal number of clusters
X1= df[["Electricity available per day(in hrs)", "Availabilty Tier"]]

wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i, init='k-means++',random_state=0)
    kmeans.fit(X1)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()

plt.scatter(df["Electricity available per day(in hrs)"], df["Availabilty Tier"])
plt.xlabel("Electricity available per day(in hrs)")
plt.ylabel("Availabilty Tier")
```

```python
#Fitting K-MEans to the dataset
kmeans=KMeans(n_clusters= 2)
y_kmeans= kmeans.fit_predict(X1)

df["Cluster"]= y_kmeans

#Visualize the clusters
df1 = df[df.Cluster==0]
df2 = df[df.Cluster==1]
#df3 = df[df.Cluster==2]

plt.scatter(df1["Electricity available per day(in hrs)"],df1["Availabilty Tier"],color='green', label= "Cluster 1")
plt.scatter(df2["Electricity available per day(in hrs)"],df2["Availabilty Tier"],color='red', label= "Cluster 2")
#plt.scatter(df3["Electric bill of last month"],df3['Electrical appliances'],color='blue')

plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='purple', marker= "*", label='Centroids')

plt.title("Clusters of customers")
plt.xlabel("Electricity available per day(in hrs)")
plt.ylabel("Availabilty Tier")
plt.legend()
```
